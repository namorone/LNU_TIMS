```python
from scipy.stats import norm, uniform, expon, binom, poisson, chisquare, chi2
import numpy as np

# Ask the user for the distribution law
law = input("Enter the distribution law (normal, uniform, exponential, binomial, Poisson): ")

# Ask the user for the significance level
alpha = float(input("Enter the significance level (e.g. 0.05 for 5%): "))

# Ask the user for the sample data
sample = input("Enter the sample data (comma-separated): ")
sample = [float(x) for x in sample.split(',')]

# Estimate the distribution parameters based on the sample data, if necessary
if law == 'normal':
    mu_hat, sigma_hat = norm.fit(sample)
    dist = norm(loc=mu_hat, scale=sigma_hat)
elif law == 'uniform':
    a_hat, b_hat = uniform.fit(sample)
    dist = uniform(loc=a_hat, scale=b_hat-a_hat)
elif law == 'exponential':
    lambda_hat = expon.fit(sample)[1]
    dist = expon(scale=1/lambda_hat)
elif law == 'binomial':
    n_hat, p_hat = binom.fit(sample)[:2]
    dist = binom(n=n_hat, p=p_hat)
elif law == 'poisson':
    mu_hat = poisson.fit(sample)[0]
    dist = poisson(mu=mu_hat)
else:
    print("Invalid distribution law")
    exit()

bins = 'auto'
obs_freq, edges = np.histogram(sample, bins=bins)

# Get the expected frequencies for each category based on the distribution
exp_freq = np.diff(dist.cdf(edges))
exp_freq *= len(sample)


# Calculate the chi-squared value and degrees of freedom
chi2_value = sum((obs_freq - exp_freq)**2 / exp_freq)
df = len(exp_freq) - 1

# Find the p-value using the CDF of the chi-squared distribution
p_value = 1 - chi2.cdf(chi2_value, df)

# Check if the null hypothesis is rejected

if p_value < alpha:
    print(f"The null hypothesis is rejected at the {alpha:.2f} level of significance")
else:
    print(f"The null hypothesis is not rejected at the {alpha:.2f} level of significance")
```
